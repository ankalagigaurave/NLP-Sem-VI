```
import nltk
nltk.download()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from nltk.corpus import stopwords
from string import punctuation
from nltk.tokenize import word_tokenize
from nltk.stem import LancasterStemmer
```

```
df=pd.read_csv('/content/spam.csv',encoding='ISO-8859-1')
df
```

```
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'], axis='columns', inplace=True)
```

```
df
```

```
df['v1']=df['v1'].replace(['ham','spam'],[0,1])
```

```
df['v1']
```

```
import re
ls=LancasterStemmer()

corpus_stem=[]
for i in range(len(df)):
    review=re.sub('[^a-zA_Z]',' ',df['v2'][i])
    review=review.lower()
    review=review.split()
    review=[ls.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus_stem.append(review)
    
corpus_stem
```

```
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()

corpus_lem=[]
for i in range(len(df)):
    review=re.sub('[^a-zA_Z]',' ',df['v2'][i])
    review=review.lower()
    review=review.split()
    review=[lem.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus_lem.append(review)
    
corpus_lem
```

```
from sklearn.feature_extraction.text import CountVectorizer
CV=CountVectorizer(max_features=5000)
x_lem=CV.fit_transform(corpus_lem).toarray()
x_lem
```

```
x_stem=CV.fit_transform(corpus_stem).toarray()
x_stem
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
tf_stem=TfidfVectorizer()
x_tf_stem=tf_stem.fit_transform(corpus_stem)
print(x_tf_stem.shape)
```

```
tf_lem=TfidfVectorizer()
x_tf_lem=tf_stem.fit_transform(corpus_lem)
print(x_tf_lem.shape)
```

```
y=df['v1']
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(x_stem,y,test_size=0.2)
```

```
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB().fit(X_train,y_train)
ypred_mnb=mnb.predict(X_test)
```

```
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,ypred_mnb))
print(confusion_matrix(y_test,ypred_mnb))
```

```
X_train, X_test, y_train, y_test= train_test_split(x_lem,y,test_size=0.2)
mnb=MultinomialNB().fit(X_train,y_train)
ypred_mnb_lem=mnb.predict(X_test)
print(classification_report(y_test,ypred_mnb_lem))
print(confusion_matrix(y_test,ypred_mnb_lem))
```

```
X_train, X_test, y_train, y_test= train_test_split(x_tf_lem,y,test_size=0.2)
mnb=MultinomialNB().fit(X_train,y_train)
ypred_mnb_tf_lem=mnb.predict(X_test)
print(classification_report(y_test,ypred_mnb_tf_lem))
print(confusion_matrix(y_test,ypred_mnb_tf_lem))
```

```
X_train, X_test, y_train, y_test= train_test_split(x_tf_stem,y,test_size=0.2)
mnb=MultinomialNB().fit(X_train,y_train)
ypred_mnb_tf_stem=mnb.predict(X_test)
print(classification_report(y_test,ypred_mnb_tf_stem))
print(confusion_matrix(y_test,ypred_mnb_tf_stem))
```

```
import imblearn
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler()
x, y = rus.fit_resample(x_lem, y)
X_train, X_test, y_train, y_test= train_test_split(x,y,test_size=0.2)
mnb=MultinomialNB().fit(X_train,y_train)
ypred_mnb_lem=mnb.predict(X_test)
print(classification_report(y_test,ypred_mnb_lem))
print(confusion_matrix(y_test,ypred_mnb_lem))
```
